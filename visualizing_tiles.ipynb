## **Moulin Tile Sampling and Visualization using 1024 x 1024 tile arrays**
## Objectives - 
1. Detect tiles containing moulins from binary mask arrays, aggregate corresponding tile indices
2. sample 20 random tiles 

**Loading folder names**

```python
import numpy as np
import gcsfs
```

```python
lake_tiles_path = "gs://yao_scratch/anagha/lake_tiles"
fs = gcsfs.GCSFileSystem()
folder_paths = fs.ls(lake_tiles_path)
folder_paths = sorted([path for path in folder_paths])[1:]
```

**Loading and storing arrays**

```python
def load_image_array(folder):
    image_tile_path = "gs://"+ folder + '/image_tiles_1024.npy'
    mask_tile_path = "gs://" + folder + '/mask_tiles_1024.npy'
    try:
        with fs.open(image_tile_path, 'rb') as image_array, fs.open(mask_tile_path, 'rb') as mask_array:
            image_tile_1024 = np.load(image_array)
            mask_tile_1024 = np.load(mask_array)
            # verifying image array type
            if type(image_tile_1024) != np.ndarray or type(mask_tile_1024) != np.ndarray:
                raise TypeError("not a numpy array")
            image_name = folder.rstrip('/').split('/')[-1]
            #loading 100 tiles for test
            image_tile_1024 = image_tile_1024[:100,...]
            mask_tile_1024 = mask_tile_1024[:100,...]            
            print(f"Loaded: {image_name}")
            return image_tile_1024,mask_tile_1024
    except FileNotFoundError:
        print(f"Missing image in: {folder}")
    except Exception as error:
        print(f"Error loading {folder}, {error}")
```

**Helper function to get moulin indices and plot sample tiles**

```python
import random 
import math 
import matplotlib.pyplot as plt

#random sampling function
def sample_moulin_tiles(img_array, sample_size=20):
    moulin_indices = [i for i in range(img_array.shape[0]) if img_array[i,0,...].sum() > 100]
    # moulin count sanity check
    print([i[0].sum() for i in img_array[moulin_indices]])
    return random.sample(moulin_indices, min(sample_size, len(moulin_indices)))
```

**Wrapper function**

```python
def final_subplot(folder, sample_size=20):
    # load data, sample moulin tiles, and plot them.
    try:
        image_array, mask_array = load_image_array(folder)
        #check array shape
        print("Shape of image array",image_array.shape)
        print("Shape of mask array",mask_array.shape)
    except Exception as error:
        print(f" Error loading {folder} - {error}")
        return
    sampled_indices = sample_moulin_tiles(mask_array, sample_size)
    image_name = folder.rstrip('/').split('/')[-1]

    fig, axes = plt.subplots(len(sampled_indices), 4, figsize=(12, 3 * len(sampled_indices)))
    
    #enumerate to get tile number + tile
    for row, tile in enumerate(sampled_indices):
        img_tile = image_array[tile, 0,...]
        # change plot color scheme to gray
        axes[row][0].imshow(img_tile, cmap='gray')
        axes[row][0].set_title("Image")

        # Moulin mask
        moulin_tile = mask_array[tile, 0,...]
        axes[row][1].imshow(moulin_tile, cmap='gray')
        axes[row][1].set_title("Moulin Mask")
        
        # Water mask
        water_tile = mask_array[tile, 1,...]
        axes[row][2].imshow(water_tile, cmap='gray')
        axes[row][2].set_title("Water Mask")
        
        # Fracture mask
        fracture_tile = mask_array[tile, 2,...]
        axes[row][3].imshow(fracture_tile, cmap='gray')
        axes[row][3].set_title("Fracture Mask")
    fig.suptitle(f"{image_name} - tile samples")
    #make sure the plots don't overlap
    plt.tight_layout()
    plt.show()
```

```python
final_subplot(folder_paths[2])
```

```python
import gcsfs
import rasterio
fs = gcsfs.GCSFileSystem()
base = "yao_scratch/anagha/images"
files = fs.find(base)

images = [f"gs://{img}" for img in files if img.endswith(".tif")]

for img in images:
    try:
        with rasterio.Env():
            with rasterio.open(img) as src:
                image_type = "Monochrome" if "P1BS" in img else "Multiband" if "M1BS" in img else "Other"
                if image_type in ("Monochrome","Multiband"):
                    print(f"{image_type}: {img.rstrip('/').split('/')[-2]}\n Size: {src.width} x {src.height} = {src.width * src.height}, Bands: {src.count}\n")
    except Exception as e:
        print(f"Error reading {img}: {e}")

```
